```
import numpy as np
import matplotlib.pyplot as plt

#dark theme
plt.rc_context({'xtick.color':'w', 'ytick.color':'w', 'text.color':'w', 'axes.labelcolor':'w'})

seed=1234
np.random.seed(seed)
```

```
x = np.linspace(0, 30, 11)
x.shape, x
```

```
y = x ** 2
y
```

```
plt.title("Demo")
plt.xlabel("x label")
plt.ylabel("y label")
plt.plot(x, y, 'gs')
plt.show()
```

```
ys = 200 + np.random.rand(100)
ys.shape
```

```
xs = [x for x in range(len(ys))]
```

```
plt.plot(xs,ys,'--')
plt.fill_between(xs, ys, 200, where=(ys > 195), facecolor='g', alpha=0.6)
plt.show()
```
